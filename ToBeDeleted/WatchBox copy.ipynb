```python
# Add Image Path assuming that the file name is given in the format MODELNUMBER.PNG
base_path = "/Users/mohjaiswal/Desktop/WatchBox/Pictures/"

# Function to create image path
def create_image_path(model_number):
    # Remove spaces and add file extension
    filename = model_number.strip() + ".png"
    return base_path + filename

# Add image paths to the DataFrame
watches_df['Image Path'] = watches_df['Model Number'].apply(create_image_path)

# Display the updated DataFrame to check the Image Path
watches_df
```

```python
# Function to display DataFrame with image previews
def display_df_with_images(watches_df, image_path_column):
    # Create a copy to modify without changing original data
    final_watches_df = watches_df.copy()

    # Convert image paths to HTML <img> tags
    final_watches_df[image_path_column] = final_watches_df[image_path_column].apply(
        lambda x: f'<img src="{x}" width="100" >' if x else '')

    # Move the image column to the first position
    cols = [image_path_column] + [col for col in final_watches_df if col != image_path_column]
    final_watches_df = final_watches_df[cols]

    # Convert DataFrame to HTML and display
    display(HTML(final_watches_df.to_html(escape=False)))

# Assuming your DataFrame is df and the image path column is 'Image Path'
display_df_with_images(watches_df, 'Image Path')
```

```python
# Assuming watches_df is your DataFrame
df = pd.DataFrame(watch_data)

# Create a dropdown widget for Brand selection
dropdown = widgets.Dropdown(
    options = df['Brand'].unique(),
    value = df['Brand'].unique()[0],
    description = 'Brand:',
    disabled = False,
)

# Function to update DataFrame based on dropdown
def dropdown_eventhandler(change):
    clear_output(wait=True)
    display(dropdown)
    filtered_df = df[df['Brand'] == change.new]
    display(filtered_df)

# Display the dropdown and DataFrame on selection
dropdown.observe(dropdown_eventhandler, names='value')
display(dropdown)
```
